# **-----------------------------Nettoyage---------------------------**

```python
import pandas as pd

# Parameters
param_nrows = None  # Set to None to load all rows, or specify a number for a subset
chunksize = 16300000

# Initialize an empty list to hold processed chunks
chunks = []

# Process the dataset in chunks
for chunk in pd.read_csv('https://datasets.imdbws.com/title.akas.tsv.gz', sep='\t', compression='gzip', na_values=["\\N"], low_memory=True, nrows=param_nrows, chunksize=chunksize):
    # Fill missing values in the language column with the values from the region column
    chunk['language'].fillna(chunk['region'], inplace=True)
    # Convert the 'language' column to lowercase
    chunk['language'] = chunk['language'].str.lower()
    # Filter the rows where 'isOriginal' is 1 or 'language' is 'fr'
    filtered_chunk = chunk[(chunk['isOriginalTitle'] == 1) | (chunk['language'] == 'fr')]
    # Drop the columns 'types' and 'attributes'
    filtered_chunk.drop(columns=['types', 'attributes'], inplace=True)
    # Append the filtered chunk to the list
    chunks.append(filtered_chunk)

# Concatenate all filtered chunks into a single DataFrame
title_akas_cleaned = pd.concat(chunks, ignore_index=True)

# Verify the changes
print(title_akas_cleaned.info())
display(title_akas_cleaned.head())
```

```python
# Save the cleaned DataFrame to a CSV file
title_akas_cleaned.to_csv('title_akas_cleaned.csv', index=False)

# Verify the changes
print(title_akas_cleaned.info())
display(title_akas_cleaned.head())
```
